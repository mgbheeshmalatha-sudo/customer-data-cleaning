```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, MinMaxScaler, StandardScaler

# Initial Dataset
df = pd.DataFrame({
    'customer_id': [1, 2, 3, 4, 2, 5],
    'age':         [25, np.nan, 35, np.nan, np.nan, 40],
    'city':        ['Mumbai', 'Delhi', np.nan, 'Mumbai', 'Delhi', np.nan],
    'gender':      ['Male', 'Female', 'Female', 'Male', 'Female', 'Male'],
    'annual_income': [40000, 80000, 60000, 120000, 80000, 95000]
})

# --- Task 1: Clean the Data ---

# Impute missing values
df['age'] = df['age'].fillna(df['age'].median())
df['city'] = df['city'].fillna(df['city'].mode()[0])

# Remove duplicate rows
df = df.drop_duplicates()

# --- Task 2: Encode Categorical Columns ---

# Label Encoding for Gender
le = LabelEncoder()
df['gender'] = le.fit_transform(df['gender'])

# One-Hot Encoding for City
df = pd.get_dummies(df, columns=['city'], prefix='city')

print("--- Cleaned and Encoded Dataframe ---")
print(df)
print("\n")

# --- Task 3: Scale Numerical Features ---

# Prepare subsets for scaling
features_to_scale = ['age', 'annual_income']

# Min-Max Scaling
mms = MinMaxScaler()
df_minmax = df.copy()
df_minmax[features_to_scale] = mms.fit_transform(df[features_to_scale])

# Standardization
ss = StandardScaler()
df_std = df.copy()
df_std[features_to_scale] = ss.fit_transform(df[features_to_scale])

print("--- Min-Max Scaled Data ---")
print(df_minmax[features_to_scale])
print("\n--- Standardized Data ---")
print(df_std[features_to_scale])
```
